```python
#生成数据集
from sklearn.datasets import make_regression
#生成特征量为1，噪音为50的数据集，特征量为1，50个噪音，创造一些线性回归的数据点
X,y=make_regression(n_features=1,n_informative=1,noise=50,random_state=8)
#用散点集对数据点可视化
plt.scatter(X,y,c='orange',edgecolor='k')
plt.show()
```

```python
#导入用于回归的KNN模型
from sklearn.neighbors import KNeighborsRegressor
reg=KNeighborsRegressor()

#使用KNN模型拟合数据
reg.fit(X,y)
#把预测结果用图像可视化
z=np.linspace(-3,3,200).reshape(-1,1)
plt.scatter(X,y,c='orange',edgecolor='k')
plt.plot(z,reg.predict(z),c='k',linewidth=3)

#向图像添加标题
plt.title('KNN Regressor')

print('\n')
print("运行结果：")
print("==========================")
print('模型评分:{:.2f}'.format(reg.score(X,y)))
print("==========================")
print('\n')


plt.show()
```

```python
#一个K临近算法的应用
#加载数据集及分析数据集，红酒数据集是一个bunch对象，dict数据类型
from sklearn.datasets import load_wine

wine_dataset=load_wine()
print('\n')
print('the result is:')
print('==============')
print("红酒数据集中的键:\n{}".format(wine_dataset.keys()))
print('==============')
print(wine_dataset.target[[10,80,140]])
print('==============')
print(wine_dataset.target_names)
print('==============')
#样本数据的数据量和变量可以使用shape获得
print("数据概况:\n{}".format(wine_dataset['data'].shape))

#数据集的细节可以使用DESCR

print(wine_dataset['DESCR'])
```

```python
#导入数据拆分工具
from sklearn.model_selection import train_test_split
#将数据拆分为训练和测试数据集
X_train,X_test,y_train,y_test=train_test_split(wine_dataset['data'],wine_dataset['target'],random_state=0)
print('=======================')
#训练数据集集中特征向量形态
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
print('X_train shape:{}'.format(X_train.shape))
#测试数据集集中特征向量形态
print('X_test shape:{}'.format(X_test.shape))
#训练数据集集中目标形态
print('y_train target shape:{}'.format(y_train.shape))
#测试数据集集中目标形态
print('y_test target shape:{}'.format(y_test.shape))
print('=======================')

```

```python
#上一步建立了数据集后，这一步开始建立模型
#导入KNN分类模型
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
#制定模型的n_neighbor参数设定为1
knn=KNeighborsClassifier(n_neighbors=1)

#使用knn中的fit（拟合）方法进行建模，使用上一步的样本数据，X_train和y_train
print('==============')
knn.fit(X_train,y_train)
print('拟合结果：\n')
print(knn)
print('==============')

#使用测试数据对模型打分
print('score for model:{:.2f}'.format(knn.score(X_test,y_test)))

print('==============')


#建立好模型后我们可以使用模型对新样品数据进行预测，但是在预测前先使用测试数据集对模型进行打分

#获取需要评定的样品特征，使用模型对新样品进行类型预测
X_new=np.array([[13.2,2.77,2.51,18.5,96.6,1.04,2.55,0.57,1.47,6.2,1.05,3.33,820]])


prediction=knn.predict(X_new)

print('\n')

print('result is {}'.format(wine_dataset['target_names'][prediction]))
```
