```python
import numpy as np
import matplotlib.pyplot as plt

X=np.linspace(-5,5,100)

y=0.5*X+3

plt.plot(X,y,c='orange')

plt.title('Straight Line')

plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression 
X=[[1],[4]]
y=[3,5]

#使用线性模型拟合这两个点
lr=LinearRegression().fit(X,y)

#画出两个点和直线的模型
z=np.linspace(0,5,20)

plt.scatter(X,y,s=80)
plt.plot(z,lr.predict(z.reshape(-1,1)),c='k')

#设定主题为straight line
plt.title('Straight Line')

#=将图片展示出来
plt.show()

#确定过这两点的直线方程

print('y={:.3f}'.format(lr.coef_[0]),'x','+ {:.3f}'.format(lr.intercept_))




#设想一下如果是三个点尼
X2=[[1],[4],[3]]
y2=[3,5,3]

#使用线性模型拟合这三个点
lr=LinearRegression().fit(X2,y2)
z2=np.linspace(0,5,20)
plt.scatter(X2,y2,s=80)
plt.plot(z,lr.predict(z.reshape(-1,1)),c='k')
plt.title('Straight Line2')
plt.show()

#计算模型画出的直线方程，coef_和intercept_分别是y=w*x+b中的w和b
print('y={:.3f}'.format(lr.coef_[0]),'x','+ {:.3f}'.format(lr.intercept_))




```

```python
#线性回归即普通最小二乘法，y的预测值与其真实值的平方差最小的时候 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression


X,y=make_regression(n_samples=100,n_features=2,n_informative=2,random_state=38)
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=8)

#拟合
lr=LinearRegression().fit(X_train,y_train)

#使用线性回归算出，w，b的值,
print('===============')

print('lr.corf_:{}'.format(lr.coef_[:]))
print('lr.intercept_:{}'.format(lr.intercept_))

print('===============')
#得到模型后，给模型打分
print('得到模型后，给模型打分')
print('score of train data :{:.2f}'.format(lr.score(X_train,y_train)))
print('score of test data :{:.2f}'.format(lr.score(X_test,y_test)))

#因为数据集没有设置如noise参数所以分数比较高
```

```python
#可以看一个线性回归在比较贴近真实数据里表现的例子
from sklearn.datasets import load_diabetes

diabetes_dataset=load_diabetes()

X,y=load_diabetes().data,load_diabetes().target

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=8)

#使用线性回归模型拟合
lr=LinearRegression().fit(X_train,y_train)

print('===============')
print('\n')
#得到模型后，给模型打分
print('score of train data :{:.2f}'.format(lr.score(X_train,y_train)))
print('score of test data :{:.2f}'.format(lr.score(X_test,y_test)))


#看一下数据特征明细
print(diabetes_dataset['DESCR'])

```
